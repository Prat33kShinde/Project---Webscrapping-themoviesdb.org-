# 15. Classification of genres 

```python

def total_count(column):
    total_count = movies[column].str.cat(sep = ',')
    total_count = pd.Series(total_count.split(','))
    total_v = total_count.value_counts(ascending = False).rename_axis('Name').reset_index(name='Counts')
    return total_v
total_v = total_count('Genres')
total_v
```

We have successfully classified genres and get total no. of values of each genres.




# 16. Top 10 most popular genres?
